```python
from keras.layers import Dense, Dropout, LSTM, Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
import pandas as pd
import numpy as np


def load_data(test_split = 0.2):
    print ('Loading data...')
    input = pd.read_csv('./data_5.txt',sep=",", header=None, dtype="float").values
    x_shape = input.shape[0]
    y_shape = input.shape[1]
    
    boundary = (int)(x_shape * 0.7)
    X_train = input[0: boundary, 0:y_shape - 1]
    y_train = input[0: boundary, -1]
    X_test = input[boundary:x_shape, 0:y_shape - 1]
    y_test = input[boundary:x_shape, -1]

    print("Data is loaded")
    return pad_sequences(X_train), y_train, pad_sequences(X_test), y_test


def create_model(input_length):
    print ('Creating model...')
    model = Sequential()
    #model.add(Embedding(input_dim = 188, output_dim = 50, input_length = input_length))
    model.add(LSTM(activation="relu", return_sequences=True, units=256, recurrent_activation="hard_sigmoid"))
             
    model.add(Dropout(0.5))
    model.add(LSTM(activation="relu", units=256, recurrent_activation="hard_sigmoid"))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    print ('Compiling...')
    model.compile(loss='binary_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model
```

```python

X_train, y_train, X_test, y_test = load_data()
```

```python

model = create_model(len(X_train[0]))

print ('Fitting model...')
hist = model.fit(X_train, y_train, batch_size=5, epochs=2, validation_split = 0.1, verbose = 1)
model.save('trained_model.h5')
score, acc = model.evaluate(X_test, y_test, batch_size=1)
print('Test score:', score)
print('Test accuracy:', acc)
```
